### Average duration of movies (min)

```python
#### This function is used to extract duration of movies in integer minutes from string

def remove_min(x):
    a,b = x.split()
    return int(a)


```

```python
movies = df[df['Category'] == "Movie"]
movies['Duration']=movies['Duration'].apply(remove_min)
movies.head(5)
```

### Distribution of movies

```python
sns.histplot(data=movies, x="Duration")
```
